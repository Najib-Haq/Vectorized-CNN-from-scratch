```python
import cv2

path = "numta/training-a/a00003.png"
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# convert to grayscale and threshold
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
(_, binary) = cv2.threshold(gray, 255//2, 255, cv2.THRESH_OTSU)
binary = 255-binary

# apply erosion + dilation to remove noise
kernel = np.ones((5,5),np.uint8)
img_opening = cv2.erode(cv2.dilate(binary,kernel,iterations = 1), kernel,iterations = 1)
# plt.figure()
# plt.imshow(img_opening, cmap='gray')

# get bounding box
contours, _ = cv2.findContours(img_opening, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours = sorted(contours, key=lambda x: cv2.contourArea(x))[-1]
bounding_box = cv2.boundingRect(contours)
    # print(bounding_box, image.shape)
bb_image = image[bounding_box[1]:bounding_box[1]+bounding_box[3], bounding_box[0]:bounding_box[0]+bounding_box[2], :]
```

```python
no_contours =contours.shape[0]

# choose random contours 
random_contours = np.random.randint(0, no_contours, 5)

c_image = image.copy()

height, width, _ = image.shape
cutout_height, cutout_width = height*0.02, width*0.02
for point in contours[random_contours]:
    # print(point, point.shape)
    x, y = point[0][0], point[0][1]
    # add a rectangle to the image
    # cv2.rectangle(c_image, (int(x-cutout_width), int(y-cutout_height)), (int(x+cutout_width), int(y+cutout_height)), (0, 255, 0), 2)

    # cutout the image
    # also check whether the cutout is within the image
    if (x-cutout_width) > 0 and (x+cutout_width) < width and (y-cutout_height) > 0 and (y+cutout_height) < height:
        c_image[int(y-cutout_height):int(y+cutout_height), int(x-cutout_width):int(x+cutout_width), :] = 255
        # plt.figure()
        # plt.imshow(cutout)
```

```python
opening_img = cv2.morphologyEx(image, cv2.MORPH_OPEN, np.ones((5, 5), np.uint8), iterations=1)

bounding_box, contours = get_number_bb(image, True)
```

```python
image = 255 - image
dilate_image = cv2.dilate(image, np.ones((5, 5), np.uint8), iterations = 1)
```

```python

final_image = get_contour_cutout(dilate_image, contours, 5)
bounded_image = image[bounding_box[1]:bounding_box[1]+bounding_box[3], bounding_box[0]:bounding_box[0]+bounding_box[2], :]
final_image = final_image[bounding_box[1]:bounding_box[1]+bounding_box[3], bounding_box[0]:bounding_box[0]+bounding_box[2], :]

contour_image = get_contour_cutout(image, contours, 5)

final_image = cv2.resize(final_image, (48, 48), interpolation = cv2.INTER_AREA)
bounded_image = cv2.resize(bounded_image, (48, 48), interpolation = cv2.INTER_AREA)
            
```

```python
path1 = "numta/training-a/a00006.png"
path2 = "numta/training-a/a00007.png"


import cv2
aimage = cv2.imread(path1)
aimage = cv2.cvtColor(aimage, cv2.COLOR_BGR2RGB)
aorig_image = aimage.copy()

aimage2 = cv2.imread(path2)
aimage2 = cv2.cvtColor(aimage2, cv2.COLOR_BGR2RGB)

aopening_img = cv2.morphologyEx(aimage, cv2.MORPH_OPEN, np.ones((5, 5), np.uint8), iterations=1)

bounding_box, contours = get_number_bb(aimage, True)
aimage = 255 - aimage
adilate_image = cv2.dilate(aimage, np.ones((5, 5), np.uint8), iterations = 1)

afinal_image = get_contour_cutout(adilate_image, contours, 5)
abounded_image = aimage[bounding_box[1]:bounding_box[1]+bounding_box[3], bounding_box[0]:bounding_box[0]+bounding_box[2], :]
afinal_image = afinal_image[bounding_box[1]:bounding_box[1]+bounding_box[3], bounding_box[0]:bounding_box[0]+bounding_box[2], :]

acontour_image = get_contour_cutout(aimage, contours, 5)

afinal_image = cv2.resize(afinal_image, (48, 48), interpolation = cv2.INTER_AREA)
abounded_image = cv2.resize(abounded_image, (48, 48), interpolation = cv2.INTER_AREA)
            
```

```python
def visualize_training(loggings, save_dir):
    fig, ax = plt.subplots(3, 1, figsize=(15, 10))
    ax[0].plot(loggings['epoch'], loggings['train_loss'], label='train')
    ax[0].plot(loggings['epoch'], loggings['val_loss'], label='val')
    ax[0].set_title('a) Loss')
    ax[0].legend()
    ax[0].grid()
    
    # ax[0].set_xlabel('Epoch')

    ax[1].plot(loggings['epoch'], loggings['train_acc'], label='train')
    ax[1].plot(loggings['epoch'], loggings['val_acc'], label='val')
    ax[1].set_title('b) Accuracy')
    ax[1].legend()
    ax[1].grid()
    
    # ax[1].set_xlabel('Epoch')

    ax[2].plot(loggings['epoch'], loggings['train_f1'], label='train')
    ax[2].plot(loggings['epoch'], loggings['val_f1'], label='val')
    ax[2].set_title('c) F1 Score')
    ax[2].legend()
    ax[2].grid()
    ax[2].set_xlabel('All X axis->Epoch')

    plt.legend()

    plt.savefig(f'{save_dir}/output.png', bbox_inches='tight')

visualize_training(df, '1705044')
```
